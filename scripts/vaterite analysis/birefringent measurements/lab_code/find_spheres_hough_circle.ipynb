```python
plt.figure(figsize=(20, 10))

# Detect two radii
hough_res = hough_circle(edges, hough_radii)

num_peaks = 25

centers = []
accums = []
radii = []

for radius, h in zip(hough_radii, hough_res):
    # For each radius, extract two circles
    peaks = peak_local_max(h, num_peaks=num_peaks)
    centers.extend(peaks)
    accums.extend(h[peaks[:, 0], peaks[:, 1]])
    radii.extend([radius] * num_peaks)

# Draw the most prominent 5 circles
image2 = gray2rgb(image)
for idx in np.argsort(accums)[::-1][:]:
    center_x, center_y = centers[idx]
    radius = int(radii[idx])
    cx, cy = circle_perimeter(center_y, center_x, radius)
    cy = cy[cx < nk]
    cx = cx[cx < nk]
    cx = cx[cy < nj]
    cy = cy[cy < nj]
    cy = cy[cx > -1]
    cx = cx[cx > -1]
    cx = cx[cy > -1]
    cy = cy[cy > -1]
    image2[cy, cx] = (220, 20, 20)

plt.imshow(image2, cmap=plt.cm.gray)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
# Draw all the circles
image2 = gray2rgb(image)
spheres = np.zeros((nj, nk), dtype=bool)
for idx in np.argsort(accums)[::-1][:]:
    center_x, center_y = centers[idx]
    radius = int(radii[idx])
    cx, cy = circle(center_y, center_x, radius)
    cy = cy[cx < nk]
    cx = cx[cx < nk]
    cx = cx[cy < nj]
    cy = cy[cy < nj]
    spheres[cy, cx] = 1
    image2[cy, cx] = (220, 20, 20)

plt.imshow(image2, cmap=plt.cm.gray)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.imshow(spheres)
plt.colorbar()
plt.show()
```

```python
background = np.bitwise_xor(np.ones((nj, nk), dtype=bool), spheres)
plt.figure(figsize=(10,5))
plt.imshow(background)
plt.colorbar()
plt.show()
```

```python
np.savez(path_name + 'sphere_locations.npz', spheres=spheres, background=background)
```

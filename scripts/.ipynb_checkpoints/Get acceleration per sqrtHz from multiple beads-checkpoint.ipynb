```python
import os, fnmatch, glob, h5py
from matplotlib.mlab import psd as get_psd_and_freqs
import numpy as np
from correlation import getGainAndACamp, get_sphere_mass, charge_of_electron, get_num_electrons
import matplotlib.pyplot as plt
import bead_util as bu

distance_between_plates = 0.001 # meters
```

```python
calib_dirs = []
list_of_pics_for_adam = []
for root, dirnames, filenames in os.walk('/data/'):
    for dirname in fnmatch.filter(dirnames, '*calibration*'):
        for filename in fnmatch.filter(filenames, '*.tif'):
            if root[0:12]=='/data/201707':
                calib_dirs.append(root)
                list_of_pics_for_adam.append(root+'/'+filename)
#calib_dirs, list_of_pics_for_adam = zip(*sorted(zip(calib_dirs, list_of_pics_for_adam)))
calib_dirs = sorted(list(set(calib_dirs)))
list_of_pics_for_adam = sorted(list(set(list_of_pics_for_adam)))

# Adam has returned results from the pictures:
sphere_diameters = [14.3579, 16.1780, 16.9869, 18.2002]
sphere_masses = map(get_sphere_mass, sphere_diameters)
num_electrons = map(get_num_electrons, sphere_diameters)

for calib_dir, pic, m in zip(calib_dirs, list_of_pics_for_adam, sphere_masses):
    print calib_dir, '        ', pic, '        ', m, ' kg'
```

```python
NFFT = 2**17

def getdata(fname, calibration=False):
	gain, acamp = getGainAndACamp(fname, need_ACamp=calibration)
	#print "Opening file: ", fname
	f = h5py.File(fname,'r')
	dset = f['beads/data/pos_data']
	dat = np.transpose(dset) * 10./(2.**15 - 1.)
	x_data = dat[:, bu.xi]/gain
	Fs = dset.attrs['Fsamp']
	xpsd, freqs = get_psd_and_freqs(x_data-np.mean(x_data), Fs = Fs, NFFT = NFFT)
	return freqs, np.array(xpsd), acamp
```

```python
temp = ['auto_xyzcool_G5_att.h5',
        '9_2E-6mbar_xyzcool_G5_att.h5',
        '5_0E-6mbar_xyzcool_G5_att.h5',
        '9E-7mbar_xyzcool_G5_att.h5']
```

```python
def get_averaged_psd(p, calibration=False):
    psd = np.array([])
    file_list = glob.glob(os.path.join(p, '*.h5'))
    freqs = np.array([])
    acamp = 0.
    for f in file_list:
        try:
            freqs, xpsd, acamp = getdata(f, calibration=calibration)
            if len(psd) != len(xpsd): psd = xpsd
            else: psd += xpsd
        except:
            continue
    psd = psd/len(file_list)
    return psd, freqs, acamp
```

```python
def calibrate(calib_dir):
    cpsd, freqs, acamp = get_averaged_psd(calib_dir, calibration=True)
    N = acamp*charge_of_electron/distance_between_plates
    binF = freqs[1] - freqs[0]
    index = np.argmax(cpsd)
    area = binF*sum(cpsd[index-2:index+2]) # V^2
    V = np.sqrt(area)
    print area, V/N, index
    return area, V/N, index
```

```python
sphere_accelerations = []
sphere_forces = []
sphere_num_electrons = []
plt.figure()
for valid_dir, m, t, en in zip(calib_dirs, sphere_masses, temp, num_electrons):
    ds = []
    for root, dirnames, filenames in os.walk(valid_dir):
        if root[0:12]=='/data/201707':
            ds.append(root)
    ds = list(set(ds))
    ds.sort(key=os.path.getmtime)
    j = max(loc for loc, val in enumerate(ds) if 'calibration' in val)
    calib_dir = ds[j]
    dir_to_analyze = ds[-1]
    
    area, c, i = calibrate(calib_dir)
    #p, f, a = get_averaged_psd(dir_to_analyze)
    f, p, a = getdata(valid_dir+'/'+t)
    plt.loglog(f, p, label=m)
    print i, f[i]
    n = np.sqrt(np.average(p[i-5:i+5])) # V/sqrtHz
    sphere_num_electrons.append(n/(area*en))
    NpsqrtHz = n/c
    sphere_forces.append(NpsqrtHz)
    apsqrtHz = NpsqrtHz/m # m/s^3/2
    sphere_accelerations.append(apsqrtHz)
plt.legend()
plt.show()
```

```python
plt.plot(sphere_diameters, sphere_accelerations)
plt.xlabel('Sphere diameters [um]')
plt.ylabel('Accelerations [m/s/s/sqrtHz]')
plt.show()
```

```python
plt.plot(sphere_diameters, sphere_forces)
plt.xlabel('Sphere diameters [um]')
plt.ylabel('Forces [N/sqrtHz]')
plt.show()
```

```python
plt.plot(sphere_diameters, sphere_num_electrons)
plt.xlabel('Sphere diameters [um]')
plt.ylabel('Fractional Charge per sqrtHz [e/sqrtHz]')
plt.show()
```

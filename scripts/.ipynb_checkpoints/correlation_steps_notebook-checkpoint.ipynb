```python
import correlation_steps as corrsteps
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib notebook

directory = "/data/20170717/bead15_15um_QWP/steps/"
calibration_path = directory + "calibration_charge/"
measurement_path = directory + "measurement_6/"

t, dc, corr = corrsteps.formData(measurement_path, calibration_path)
```

```python
plt.figure()
plt.plot(t, np.array(corr), 'o')
plt.xlabel('Time [ms]')
plt.ylabel('Correlation between drive and response [e]')
plt.title('Correlation vs time')
plt.show()
```

```python
plt.hist(corr, bins=100)
plt.show()
```

```python
corr_array = []
i = 2
while i < len(corr):
    j = i + 8
    corr_array.append(sum(np.array(corr[i:j]))/(32.))
    i = j
plt.figure()
plt.plot(corr_array)
plt.ylabel('Correlation between drive and response [e]')
plt.title('Correlation vs. Time')
plt.show()
```

```python
plt.hist(corr_array, bins=25)
plt.show()
```

```python
np.median(corr_array)
```

```python
np.std(corr_array)/np.sqrt(len(corr_array))
```

```python
%matplotlib notebook
cpsd = np.fft.fft(corr)
freqs = np.linspace(-0.010, 0.010, len(cpsd))
plt.figure()
plt.plot(freqs, abs(cpsd))
plt.show()
```
